```python
from scipy import stats
import statistics as stat

edu_PartialHighSchool = [30000, 10000, 10000, 20000, 70000, 30000, 10000, 20000, 10000, 30000]
edu_HighSchool = [20000, 30000, 60000, 20000, 40000, 60000, 60000]
edu_PartialCollege = [60000, 40000, 30000, 130000, 60000, 40000]
edu_Bachelors = [40000, 20000, 90000, 120000, 60000, 130000, 20000, 100000, 80000, 30000, 100000, 120000]
edu_Graduats = [160000, 50000, 80000, 30000, 40000, 80000, 130000, 100000, 130000, 50000, 130000]

print(f'Partial high School: {str(int(stat.mean(edu_PartialHighSchool)))}')
print(f'High School        : {str(int(stat.mean(edu_HighSchool)))}')
print(f'Partial College    : {str(int(stat.mean(edu_PartialCollege)))}')
print(f'Bachelors          : {str(int(stat.mean(edu_Bachelors)))}')
print(f'Graduate           : {str(int(stat.mean(edu_Graduats)))}')
```

```python
t, p = stats.ttest_ind(edu_PartialHighSchool, edu_HighSchool)

print(f't-value: {t}')
print(f'p-value: {p}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

sns.distplot(edu_PartialHighSchool, label="Partial High School")
sns.distplot(edu_HighSchool, label="High School")
sns.distplot(edu_PartialCollege, label="Partial College")
sns.distplot(edu_Bachelors, label="Bachelors")
sns.distplot(edu_Graduats, label="Graduate")
plt.legend()
```

```python
f, p = stats.f_oneway(edu_Bachelors, edu_Graduats, edu_HighSchool, edu_PartialCollege, edu_PartialHighSchool)

print(f'f-value: {f}')
print(f'p-value: {p}')
```

```python
import pandas as pd

df = pd.read_csv('https://www.ishelp.info/data/insurance.csv')

df
```

```python
sns.histplot(data=df, x='charges', hue='smoker', kde=True);
```

```python
smoker_y = df[df['smoker'] == 'yes']
smoker_n = df[df['smoker'] == 'no']

stats.ttest_ind(smoker_y['charges'], smoker_n['charges'])
```

```python
# feature = 'smoker'
# label = 'charges'

def anova(feature, label):

    groups = df[feature].unique()
    # print(groups)

    grouped_values = []
    for group in groups:
    #     print(group)
        grouped_values.append(df[df[feature]==group][label])

    grouped_values

    return stats.f_oneway(*grouped_values)
```

```python
anova('smoker', 'charges')
```
